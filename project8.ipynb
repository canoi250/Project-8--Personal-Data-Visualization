Part 4: How often were my favorite songs present in the playlist:

    -Odorikio by vaundy
    -Shinunoga E -Wa by Fuji Kaze
    -The Weekend by BIBI
    -Hey kids, close your eyes by AKMU
    -wet dreams by Daniela Andrade

```python
#assign all track IDs to recognizable variable name
odoriko = "spotify:track:4UeWKazLR1ZwwSVnLw9Ir9"
e_wa = "spotify:track:0o9zmvc5f3EFApU52PPIyW"
the_weekend = "spotify:track:5q3LwAHTqo9d3rET2EA9Nq"
close_your_eyes = "spotify:track:4Cgct9Vlype9cYZFW2wSnk"
wet_dreams = "spotify:track:1GjrjKqpoUiopKLmEwAbB9"
```

```python
spotify_data =  pd.read_json("OnRepeatContents.json")
spotify_data["message_contents"]
```

```python
#initialze the count variable
odoriko_count = 0
#initialze an empty list that holds each timestamp where Odoriko is present in the playlist
odoriko_timestamp = []
#loop through both columns
for timestamp, contents in zip(spotify_data["timestamp_utc"], spotify_data["message_contents"]):
    for item in contents:
        #if odoriko is in the list of songs add to the song count and add timestamp to the empty list
        if item == odoriko:
            odoriko_count += 1
            odoriko_timestamp.append(timestamp)
print("Total count:", odoriko_count)
odoriko_timestamp

vaundy_timestamps = pd.DataFrame(odoriko_timestamp)
```

```python
vaundy_timestamps
```

```python
#finds the amount of times odoriko was in the playlist every month
odoriko_month_count = vaundy_timestamps[0].dt.month.value_counts()
```

```python
odoriko_month_count
```

```python
ax = odoriko_month_count.plot(kind='bar')
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.set_title('Number of times Odoriko showed up in "OnRepeat" playlist per month')
plt.show()
```

```python
#initialize e_wa count
e_wa_count = 0
#loop through a list in message_contents column
for lst in spotify_data["message_contents"]:
    #loop through the list and checks if shinunoga E-Wa is present in the list
    for item in lst:
        if item == e_wa:
            e_wa_count = e_wa_count + 1
print(e_wa_count)
```

```python
#initialize counter variable
close_your_eyes_count = 0
#loop through a list in message_contents column
for lst in spotify_data["message_contents"]:
    #loop through the list and checks if Hey kid, close your eyes is present in the list
    for item in lst:
        if item == close_your_eyes:
            close_your_eyes_count = close_your_eyes_count + 1
print(close_your_eyes_count)
```

```python
#initialize counter variable
wet_dreams_count = 0
#loop through a list in message_contents column
for lst in spotify_data["message_contents"]:
    #loop through the list and checks if Wet Dreams is present in the list
    for item in lst:
        if item == wet_dreams:
             wet_dreams_count = wet_dreams_count + 1
print(wet_dreams_count)
```

```python
#initialize counter variable
the_weekend_count = 0
#loop through a list in message_contents column
for lst in spotify_data["message_contents"]:
    #loop through the list and checks if The Weekend is present in the list
    for item in lst:
        if item == the_weekend:
            the_weekend_count = the_weekend_count + 1
print(the_weekend_count)
```

```python
# Data of all song counts
data = {'Odoriko': odoriko_count ,
        'Shinunoga E-Wa': e_wa_count,
        'The Weekend': the_weekend_count,
        'Wet Dreams' : wet_dreams_count ,
        'Hey kid, Close Your Eyes' :close_your_eyes_count }

# Convert the data to a list of values and labels
labels = list(data.keys())
values = list(data.values())

# Plot
fig, ax = plt.subplots()
ax.bar(labels, values)
ax.set_title('Number of times a track was present in the playlist')
ax.set_xlabel('Track')
ax.set_ylabel('Count')
plt.xticks(fontsize=7)
plt.show()
```
